$$ ||e_{rel}||_p = \frac{ ||e_{abs}||_p } {||x_{true}||_p}$$

```python
# HW8.8. Relative Vector Norm Error with Floating Point
p=1
x_true = np.array([6, 5.25])
                #  110, 101.01
                #  1.10*2^2, 1.010*2^2
x_approx = np.array([6, 5])

e_abs = x_true - x_approx

norm_e_rel = LA.norm(e_abs, p)/LA.norm(x_true,p)
print(norm_e_rel)
```

```python
# HW8.11. Norm Ball
# input:
r = 2

################## Start Here ##################

import matplotlib.pyplot as plt
import numpy as np

# It may be helpful to complete this function
def norm(x, p):
    """Computes a p-norm.

    Args:
        x (ndarray): input array
        p (int or float): order of the norm

    Returns:
        (float): the p-norm of the array
    """
    # norm_x = np.linalg.norm(x, p, axis=0) # not allowed in this question
    norm_x = []
    for point in x.T:
        norm_point = (np.sum(abs(point)**p))**(1/p)
        norm_x.append(norm_point)
    norm_x = np.array(norm_x)
    return norm_x

ps = [1, 2, 5, 0.5]
phi = np.linspace(0, 2 * np.pi, 500)
unit_circle = np.array([np.cos(phi), np.sin(phi)])

plt.figure('fig_1')
# get radius r ball
norm1 = norm(unit_circle, p=1)
w = unit_circle/norm1*r
plt.plot(w[0], w[1])
fig_1 = plt.gca()
fig_1.set_aspect('equal')

plt.figure('fig_2')
# get radius r ball
norm2 = norm(unit_circle, p=2)
w = unit_circle/norm2*r
plt.plot(w[0], w[1])
fig_2 = plt.gca()
fig_2.set_aspect('equal')

plt.figure('fig_3')
# get radius r ball
norm3 = norm(unit_circle, p=5)
w = unit_circle/norm3*r
plt.plot(w[0], w[1])
fig_3 = plt.gca()
fig_3.set_aspect('equal')

plt.figure('fig_4')
# get radius r ball
norm4 = norm(unit_circle, p=0.5)
w = unit_circle/norm4*r
plt.plot(w[0], w[1])
fig_4 = plt.gca()
fig_4.set_aspect('equal')
```
